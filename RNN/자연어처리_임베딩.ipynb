## nltk

```python
import nltk
nltk.download('punkt')
text = nltk.word_tokenize('Is it possible distinguishing cats and dogs')
text
```

```python
# 품사 태깅에 대한 출력 결과
nltk.download('averaged_perceptron_tagger')
nltk.pos_tag(text)
```

```python
from konlpy.tag import Komoran
komoran = Komoran()
print(komoran.morphs('딥러닝이 쉽나요? 어렵나요?'))
```
